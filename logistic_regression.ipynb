```python
import numpy as np 
import pandas as pd 
```

```python
data=pd.read_csv("/content/SalesKaggle3.csv")
data.head(5)
```

```python
data.isnull()
```

```python
data['New_Release_Flag'].unique()
```

```python
data['File_Type'].unique()
```

```python
data.drop(['Order','MarketingType','ReleaseYear','SoldCount','SoldFlag','ReleaseNumber','File_Type'],axis=1,inplace=True)
```

```python
data.replace(to_replace='Historical',value=0,inplace=True)
data.replace(to_replace='Active',value=1,inplace=True)
```

```python
data.head(192046)
```

```python
dep_var=data['New_Release_Flag']
indep_var = data.drop(['New_Release_Flag'],axis=1)
```

```python
indep_var
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(indep_var)
print(x)
```

```python
indep_var= pd.DataFrame(x,columns=indep_var.columns)
indep_var.head(5)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=4)
pca.fit(indep_var)
principal_components = pca.transform(indep_var)
pca.explained_variance_ratio_
```

```python
print(abs( pca.components_ ))
```

```python
principaldata=data.drop(['New_Release_Flag','LowUserPrice','LowNetPrice'], axis=1)
principaldata.head(5)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(principaldata,dep_var,test_size=0.2,random_state=43)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import 
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
model1 = LogisticRegression()
model2 = RandomForestClassifier()
model3 = KNeighborsClassifier()
model4 = GaussianNB()
```

```python
model1.fit(x_train,y_train)
pred1=model1.predict(x_test)
print(accuracy_score(y_test,pred1))
```

```python
model2.fit(x_train,y_train)
pred2=model2.predict(x_test)
print(accuracy_score(y_test,pred2))
```

```python
model3.fit(x_train,y_train)
pred3=model3.predict(x_test)
print(accuracy_score(y_test,pred3))
```

```python
model4.fit(x_train,y_train)
pred4=model4.predict(x_test)
print(accuracy_score(y_test,pred4))
```
